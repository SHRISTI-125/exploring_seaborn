```python
sns.set_style("whitegrid",{
    "axes.facecolor" : "white", 
    "axes.edgecolor" : "black",
    "axes.labelcolor" : "green",
    "grid.color" : "#000000",
    "font.family" : "Arial",
    "grid.linespace" : "-",
    "axes.spines.top" : False,
    "axes.spine.right" : False
})
plt.style.use("dark_background")
sns.distplot(food_data["wheat_production"], color = "magenta")

plt.show()
```

CHAPTER : 04 DISTRIBUTION WITHIN CATEGORIES

```python
smartphone_data = pd.read_csv("smartphones.csv")
```

```python
smartphone_data
```

```python
sns.boxplot(x = "price", y = "rating", data = smartphone_data)
plt.show()
```

```python
smt_data = smartphone_data[0:10]
sns.boxplot(x = "rating", y = "price", data = smt_data)
plt.show()
```

```python
sns.violinplot(x = "rating", y = "price", data = smartphone_data[:25])
plt.show()
```

```python
sns.violinplot(x = "rating", y = "price", data = smartphone_data[:25], bw = 5, scale = 'count')
plt.show() # violinplot with bandwidth and scale
```

```python
sns.violinplot(x = "rating", y = "price", data = smartphone_data[:25],
               bw = 0.5, scale = 'count',
              palette = 'rainbow')
plt.show() 
```

```python
sns.swarmplot(x = "rating", y = "price", data = smartphone_data[:25], color = 'blue', size = 3)
plt.show() 
```

```python
sns.set(style = "whitegrid", color_codes = True)
sns.stripplot(x = "rating", y = "price", data = smartphone_data[:25], size = 6, jitter = True)
plt.show() 
```

```python
sns.swarmplot(x = "price", y = "rating", data = smartphone_data[:7], hue="model", size = 20)
plt.show() 
```

```python
sns.countplot(x = "os", data = smartphone_data[:20], palette = 'Accent')
plt.show() 
```

```python
sns.catplot(x = "price", y = "rating", data = smartphone_data[:5], row = "model", kind = "box", height = 2, aspect = 2)
plt.show()
```

```python
sns.pointplot(x = "rating", y = "price", data = smartphone_data[:20])
plt.show()
```
